```python
from transformers import pipeline
from langchain_huggingface import HuggingFacePipeline

pipe = pipeline(
    task="text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    max_new_tokens=200,
    temperature=0.2,
)
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    template="""
You are a helpful assistant.

Answer only using the transcript below.

If the transcript doesn't contain the answer, simply say "I don't know."

Transcript:
{context}

Question:
{question}

Answer:
""",
    input_variables=["context", "question"],
)
```

```python
question="is the topic of aliens disscussed in this video? if yes then what was discussed"
retriever_docs=retriever.invoke(question)
```

```python
retriever_docs
```

```python
context_text="\n\n".join(doc.page_content for doc in retriever_docs)
```

```python
context_text
```

```python
final_prompt = prompt.format(
    context=context_text,
    question=question
)
answer = llm.invoke(final_prompt)

print(answer)
```

```python
from langchain_core.runnables import RunnableParallel, RunnableLambda,RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
def format_docs(retrieved_docs):
    return "\n\n".join(doc.page_content for doc in retrieved_docs)
parllel_chain=RunnableParallel({
    'context':retriever|RunnableLambda(format_docs),
    'question':RunnablePassthrough()
})
```

```python
parllel_chain.invoke('who is Demis')
```

```python
parser=StrOutputParser()
```

```python
main_chain=parllel_chain|prompt|llm|parser
```

```python
main_chain.invoke("Can you summarize this video")
```

```python
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
documents=[
    Document(page_content="LangChain helps develpers build application easily"),
    Document(page_content="Chroma is a vector database for LLm based search "),
    Document(page_content="Embeddings convert text in to high dimensional vectors"),
    Document(page_content="OpenAI provides powerful embeddings models"),
]
```

```python
embidding_model=HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
vectorstore=Chroma.from_documents(
        documents=documents,
        embedding=embidding_model,
        collection_name="my_collection"     
)
```

```python
retriever=vectorstore.as_retriever(search_kwargs={'k':2})
```

```python
query="What is the Chroma  used for?"
results=retriever.invoke(query)
```

```python
for i, doc in enumerate(results):
    print(f"\n--Result {i+1}---")
    print(doc.page_content)
```

```python
#  we can do same from this also 
results=vectorstore.similarity_search(query,k=2)
for i, doc in enumerate(results):
    print(f"\n--Result {i+1}---")
    print(doc.page_content)
```
